z & = (2x - 3y^2)^5 \\ \\
z & = u^5 \text{ with } u = (2x - 3y^2) \\
\frac{\partial z}{\partial x} & = \frac{\partial z}{\partial u} \cdot \frac{\partial u}{\partial x} \\
& = 5u^4 \cdot 2 = 10u^4 = 10(2x - 3y^2)^4 \\ \\
\frac{\partial z}{\partial y} & = \frac{\partial z}{\partial u} \cdot \frac{\partial u}{\partial y} \\
& = 5u^4 \cdot -6y = -30u^4 \cdot y \\
& = -30y(2x -3y^2)^4
\end{align*}

(b)

\begin{align*}
z & = x^2 \cdot e^{-xy} \\ \\
z & = uv \text{ with } u = x^2, v = e^{-xy} \\
\frac{\partial z}{\partial x} & = \frac{\partial u}{\partial x} v + u \frac{\partial v}{\partial x} \\ \\
\frac{\partial u}{\partial x} & = 2x \\ \\
v & = e^t \text{ with } t = -xy \\
\frac{\partial t}{\partial x} & = -y \\
\frac{\partial v}{\partial x} & = e^{-xy} \cdot (-y) \\ \\
\frac{\partial z}{\partial x} & = 2x \cdot v + u \cdot -ye^{-xy} \\
& = 2x \cdot e^{-xy} - x^2 -ye^{-xy}
\end{align*}
